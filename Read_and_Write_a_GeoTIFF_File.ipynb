# Read and Write a GeoTIFF File
An Xarray data array can be made by reading a GeoTIFF file using the `open_rasterio()` function from the *Rioxarray* package. This then leads to opportunities for reading a time series of GeoTIFFs into a single data array, and stacking time series of GeoTIFF bands into Xarray datasets. Alternatively, Xarray data arrays can be converted to GeoTIFF for exporting to GIS applications such as QGIS.

*Note*: This method works with single-band GeoTIFFs only.

## Setup

```python
import xarray as xr
import rioxarray # For reading and writing GeoTIFFs
```

## Read a GeoTIFF File

```python
da = rioxarray.open_rasterio('Data/GeoTIFF/NDVI_2018.tiff')
da
```

## Write to a GeoTIFF File

```python
da.rio.to_raster('Data/Output/NDVI_2018_test.tiff')
```
