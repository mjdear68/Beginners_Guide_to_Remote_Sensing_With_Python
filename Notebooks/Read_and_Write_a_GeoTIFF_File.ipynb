# Read and Write a GeoTIFF File
An Xarray data array can be made by reading a GeoTIFF file using the `open_rasterio()` function from the *Rioxarray* package. The GeoTIFF file can contain a single band or multiple bands. This is useful for reading a time series of individual GeoTIFF files into a single data array which can then be stacked into a multiple band time series Xarray dataset. Alternatively, Xarray datasets can be converted to single or multiple band data arrays for exporting as GeoTIFFs. These can then be imported into GIS applications such as QGIS.

## Setup

```python
import xarray as xr
import rioxarray # For reading and writing GeoTIFFs
```

## Read a Single Band GeoTIFF File

```python
da = rioxarray.open_rasterio('../Data/GeoTIFF/Single_Band/NDVI_2018.tiff')
da
```

## Write to a GeoTIFF File

```python
da.rio.to_raster('../Output/NDVI_2018_test.tiff')
```

## Read a Multiple Band GeoTIFF File

```python
da = rioxarray.open_rasterio('../Data/GeoTIFF/Multiple_Band/MultiBand_2018.tiff')
da
```

## Write to a GeoTIFF File

```python
da.rio.to_raster('../Output/MultiBand_2018_test.tiff')
```
