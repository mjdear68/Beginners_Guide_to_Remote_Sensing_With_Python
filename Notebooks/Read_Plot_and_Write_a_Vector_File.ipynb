# Read, Plot and Write a Vector File
GeoPandas is built on the Pandas library. A GeoPandas GeoDataframe object has many of the functions and attributes of Pandas DataFrame object, plus a geometry variable that contains the geospatial data.

## Setup

```python
import geopandas as gpd # For reading and writing vector files
```

## Read the Vector File
GeoPandas can read most spatial data formats. The vector file in this example contains the boundary of the Greater Blue Mountains World Heritage area, in which our study area is situated, in ESRI shapefile format. Reading a file is achieved using the `.read_file()` function.

```python
gdf = gpd.read_file('../Data/Vector/gbm/gbm.shp')
```

```python
# Inspect the contents of the GeoDataFrame
gdf
```

## Plot
Plotting simply requires applying the `.plot()` function to the GeoDataFrame. This will create a filled plot. A boundary plot can be made by using the GeoDataFrame's `.boundary` attribute.

```python
gdf.plot()
```

```python
gdf.boundary.plot()
```

## Write to File
GeoPandas can also write to many different file formats. The example below writes to a file in the GeoJSON format.

```python
gdf.to_file("../Output/gbm.geojson", driver="GeoJSON")
```
