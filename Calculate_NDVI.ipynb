# Calculate NDVI

## Setup

```python
import xarray as xr
import rioxarray
```

## Read the netCDF File

```python
ds = xr.open_dataset('Data/netCDF/ds_BM_NP.nc')
```

## Calculate Monthly NDVI

```python
NDVI = (ds.nbart_nir_1 - ds.nbart_red)/(ds.nbart_nir_1 + ds.nbart_red)
```

```python
NDVI.sel(time=['2018-06-30','2018-12-31']).plot(col='time')
```

## Calculate Annual Median NDVI

```python
ds_yr = ds.resample(time='Y').median()
```

```python
NDVI_yr = (ds_yr.nbart_nir_1 - ds_yr.nbart_red)/(ds_yr.nbart_nir_1 + ds_yr.nbart_red)
```

```python
NDVI_yr.plot(col='time')
```
